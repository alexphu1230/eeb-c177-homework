```python
import numpy as np 
data = np.loadtxt (fname= '../week-four/inflammation-01.csv', delimiter= ",")
```

```python
print (data)
```

```python
import matplotlib.pyplot as plt 
```

```python
image = plt.imshow(data)
plt.show ()
#darker colors represent lower intensity values and brighter colors represent high intensity values 
```

```python
ave_inflammation= np.mean(data, axis=0)
ave_plot= plt.plot (ave_inflammation)
plt.xlabel('days')
plt.ylabel('inflammation score')
plt.title('patient data')
plt.show ()
```

```python
max_plot= plt.plot(np.max(data, axis=0))
plt.show
```

```python
min_plot= plt.plot(np.min(data, axis=0))
plt.show()
```

```python
fig = plt.figure(figsize=(10.0,3.0))

fig.suptitle('Patient Data', va='bottom')

sbplt1 = fig.add_subplot (1,3,1)
sbplt2= fig.add_subplot (1,3,2)
sbplt3= fig.add_subplot (1,3,3)

sbplt1.set_ylabel ('average')
sbplt1.plot(np.mean(data, axis=0), '--')

sbplt2.set_ylabel ('max')
sbplt2.set_xlabel ('time (days)')
sbplt2.plot(np.max(data,axis=0), 'o-')

sbplt3.set_ylabel ('min')
sbplt3.plot (np.min(data, axis=0), '*')

fig.tight_layout()

plt.show ()
```

```python
import csv 
import re 
```

```python
inputpath= "bee_list.txt"
```

```python
authors= []
with open (inputpath, "r") as dset: 
    data=csv.DictReader(dset, delimiter= "\t")
    for row in data:
        authors.append(row["Taxon Author"])
    header= data.fieldnames
    
print(header)
```

```python
data.fieldnames
```

```python
authors
```

```python
string = "Tadauchi, Hirashima & Matsumura, 1987"

def check_auth(string):
    notnames = re.compile('[\&\,\(\)0-9]+')
    authorlist = []
    for result in notnames.split(string):
        if result not in (" "):
            authorlist.append(result.strip())
    return (authorlist)
```

```python
mycounter = dict()

for people in authors:
    authorlist= check_auth(people)
    for author in authorlist:
        if author not in mycounter:
            mycounter[author]=0
        mycounter[author] += 1
print (mycounter)
```

```python
maxcount, maxauthor = max ((count, author) for author, count in mycounter.items())
print (maxcount)
print (maxauthor)
```
